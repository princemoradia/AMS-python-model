```python
import cv2

# Open a connection to the webcam (0 represents the default camera)
cap = cv2.VideoCapture(0)

# Check if the webcam is opened successfully
if not cap.isOpened():
    print("Error: Could not open webcam.")
    exit()
    
while True:
        
        i=0
        # Capture a single frame from the webcam
        ret, frame = cap.read()

        # Save the captured frame to an image file (you can change the filename and format)
        cv2.imwrite("jay.jpg", frame)

        # Release the webcam
        #cap.release()

        # Display a message indicating the successful capture
        print("Image captured successfully.")
        i = i+1

        # Optionally, you can also display the captured image
        cv2.imshow("Captured Image", frame)
        
        if cv2.waitKey(2) & 0xFF == ord('x'):
            break
    
    
cap.release()      
cv2.destroyAllWindows()
```
